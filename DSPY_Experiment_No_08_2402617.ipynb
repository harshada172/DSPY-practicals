**1. What is 2 tailed and  tailed T Test ?**

A T-test is a statistical hypothesis test that compares the means of two groups. Here's the difference between one-tailed and two-tailed tests:

  * Two-tailed T-test: This test is used when you want to determine if there is a significant difference between two group means in either direction (i.e., if one mean is greater or less than the other). The null hypothesis states there is no difference, and the alternative hypothesis states there is a difference.

  * One-tailed T-test: This test is used when you have a specific direction in mind for the difference between group means. For example, you might only be interested in whether group A's mean is greater than group B's mean, or whether it's less than group B's mean. The alternative hypothesis specifies the direction of the difference.

**2. Explain what is Type I and Type II errors with an example.**

* Type I Error (False Positive): This occurs when you incorrectly reject a true null hypothesis. In simpler terms, you conclude there is a significant effect or difference when there isn't one.

  Example: Imagine a new drug is being tested for a disease. The null hypothesis is that the drug has no effect. A Type I error would be concluding that the drug is effective when, in reality, it does nothing.

* Type II Error (False Negative): This occurs when you incorrectly fail to reject a false null hypothesis. In other words, you conclude there is no significant effect or difference when there actually is one.

  Example: Using the same drug example, a Type II error would be concluding that the drug is not effective when, in reality, it does work.



**3. What is power of statistical test?**

The power of a statistical test is the probability of correctly rejecting a false null hypothesis. In simpler terms, it's the likelihood that your test will detect an effect or difference when there actually is one. A test with high power is more likely to find a true effect, while a test with low power might miss a real effect (leading to a Type II error).

**4. Explain input and out put characteristics for  stats.ttest_rel function in your own words**

The scipy.stats.ttest_rel function is used to calculate the T-test for dependent (related) samples, which means you have two measurements from the same subjects or matched pairs. Here's a breakdown of its input and output characteristics:

Input Characteristics (What you provide to the function):

  * a (array_like): This is the first set of measurements. Think of it as the 'before' or 'treatment 1' data. It should be an array or list of numerical values.
  * b (array_like): This is the second set of measurements, corresponding to a. Think of it as the 'after' or 'treatment 2' data. It must have the same size as a and contain numerical values. The critical point is that each element in a is directly related to the corresponding element in b (e.g., blood pressure before a drug and after the drug for the same person).
  * axis (int or None, optional): If your input arrays (a and b) are multi-dimensional, this parameter specifies the axis along which to compute the test. By default, it computes the test along the last axis. For simple 1D arrays, you usually don't need to specify this.
 * nan_policy ({'propagate', 'raise', 'omit'}, optional): This handles how NaN (Not a Number) values are treated:
        'propagate' (default): If there are NaNs, the result will also contain NaNs.
        'raise': If there are NaNs, an error is raised.
        'omit': NaN values are ignored. This means that if a pair has a NaN, that entire pair is excluded from the calculation.
  * alternative ({'two-sided', 'less', 'greater'}, optional): This specifies the alternative hypothesis:
        'two-sided' (default): Tests if the means are different (either greater or less).
        'less': Tests if the mean of a is less than the mean of b.
        'greater': Tests if the mean of a is greater than the mean of b.

Output Characteristics (What the function returns):

  * statistic (float or array): This is the calculated t-statistic. It's a measure of the difference between the two sample means relative to the variation within the samples. A larger absolute value typically indicates a greater difference.
  * pvalue (float or array): This is the two-tailed p-value (unless you specified alternative='less' or alternative='greater'). The p-value helps you determine the statistical significance of your results. A small p-value (typically less than 0.05) suggests that the observed difference between the means is unlikely to have occurred by random chance, leading you to reject the null hypothesis.

**5. What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**

The `statsmodels` library is often used for power analysis. While `scipy.stats` is excellent for many statistical functions, `statsmodels.stats.power` provides dedicated tools for calculating the power of a test, determining required sample sizes, effect sizes, or significance levels.

```python
import statsmodels.stats.power as smp

# Create an instance of the power analysis for independent t-test
power_analysis = smp.TTestIndPower()

# Calculate the power of a test given:
#   effect_size (standardized difference between means)
#   nobs1 (number of observations in one sample)
#   alpha (significance level)
#   ratio (ratio of nobs2 / nobs1, if sample sizes are different)

# Example: Calculate power for an effect size of 0.5, 50 observations per group, and alpha=0.05
power = power_analysis.solve_power(
    effect_size=0.5, nobs1=50, alpha=0.05, ratio=1.0, alternative='two-sided'
)

print(f"Calculated Power: {power:.4f}")
```

**Output Range and Significance:**

When you solve for `power`, the output is a **probability value, ranging from 0 to 1** (inclusive).

*   **Range (0 to 1):**
    *   A power of **0** means the test has no chance of detecting a true effect (if one exists).
    *   A power of **1** means the test is guaranteed to detect a true effect (if one exists).

*   **Significance:**
    The power of a statistical test represents the **probability of correctly rejecting a false null hypothesis**. In simpler terms, it's the probability that your test will detect an effect when there actually *is* an effect to be detected.

    *   **High Power (e.g., 0.8 or higher):** This is generally desired. It indicates a high chance of finding a statistically significant result if the alternative hypothesis is true. A test with high power is less likely to commit a Type II error (false negative).
    *   **Low Power (e.g., below 0.5):** This means the test has a poor chance of detecting a true effect. If a study with low power yields a non-significant result, it's difficult to conclude that there's no effect; it might just be that the test wasn't strong enough to detect it (a high risk of Type II error).

Power analysis is crucial for study design, helping researchers determine the necessary sample size to detect a hypothesized effect with a reasonable level of confidence.
