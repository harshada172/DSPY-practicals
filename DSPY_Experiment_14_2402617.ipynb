### **Step 2: Load Dataset**

```python
from google.colab import files
import io

try:
    print("Please upload the 'mental_health.csv' file:")
    uploaded = files.upload()

    if 'mental_health.csv' in uploaded:
        df = pd.read_csv(io.BytesIO(uploaded['mental_health.csv']))
        print("Dataset 'mental_health.csv' loaded successfully.")
        df.head()
    else:
        print("Warning: 'mental_health.csv' was not among the uploaded files. Please ensure you upload the correct file.")
        df = None # Ensure df is not defined if file not found
except Exception as e:
    print(f"An error occurred during file upload or reading: {e}")
    df = None
```

The dataset is loaded using pandas. The head() function displays the first 5 rows to understand the structure of the dataset.

### **Step 3: Data Selection**

### **Step 8: Data Wrangling**
**Remove unnecessary column**

```python
df.drop('Person_ID', axis=1, inplace=True)
```

**Convert categorical data**

```python
df['Gender'] = df['Gender'].astype('category')
df['Occupation'] = df['Occupation'].astype('category')
df['Physical_Activity'] = df['Physical_Activity'].astype('category')
```
