**Array Slicing: Accessing Subarrays**

Just as we can use square brackets to access individual array elements, we can also use them to access subarrays with the slice notation, marked by the colon (:) character. The NumPy slicing syntax follows that of the standard Python list; to access a slice of an array x, use this:

If any of these are unspecified, they default to the values start=0, stop=size of dimension, step=1. We'll take a look at accessing sub-arrays in one dimension and in multiple dimensions.

**Introducing UFuncs**

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.


```python
#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

```python
#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)
```

Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

**4. Consider array object named Obj with dimensions 3*4*2 and you want to convert it to 4*6 shape, explain how will you do it? Show this using a code cell, taking an example of 34*2 np array of random int between 1 and 50.**

To reshape a NumPy array `Obj` from dimensions 3x4x2 to 4x6, we need to ensure that the total number of elements remains the same. In this case, 3 * 4 * 2 = 24 elements, and 4 * 6 = 24 elements, so reshaping is possible.

We can achieve this using the `reshape()` method. The `reshape()` method returns a new array with the same data but a different shape. The original array is not modified. You pass the new desired shape as a tuple to the `reshape()` method.

```python
import numpy as np

# Create a 3x4x2 NumPy array with random integers between 1 and 50
Obj = np.random.randint(1, 51, size=(3, 4, 2))
print("Original Array (Obj):\n", Obj)
print("Original Shape:", Obj.shape)
print("Original Number of elements:", Obj.size)

# Reshape the array to 4x6
reshaped_Obj = Obj.reshape(4, 6)
print("\nReshaped Array:\n", reshaped_Obj)
print("New Shape:", reshaped_Obj.shape)
print("New Number of elements:", reshaped_Obj.size)
```

**Conclusion:**

After modifying an element in `subObjCpy`, you'll observe that the corresponding element in `reshaped_Obj` (and consequently in the original `Obj` from which `reshaped_Obj` was derived) has also changed. This demonstrates that NumPy array slices, by default, return *views* of the original array, not independent copies. **7. If you were to filter sub array view from above Obj object and want to apply modification in original and/or sub obj change respective element in other how will you achieve it?**

As established in the previous answer, NumPy array slices (which often includes 'filtered sub array views' if created directly through slicing or certain indexing) return *views* of the original array's data. This means that if you modify an element in the sub-array view, that change will automatically be reflected in the original array, and vice-versa.

To ensure modifications apply to both the view and the original, you simply perform the modification on the view (or the original, it's the same underlying data).

Let's consider our `reshaped_Obj` again and create a filtered view based on a condition, then modify an element in that filtered view.

```python
print("Original reshaped_Obj before filtering and modification:\n", reshaped_Obj)

# Create a filtered sub-array view where values are greater than 40
# This uses boolean indexing, which also returns a view for simple assignments
filtered_view = reshaped_Obj[reshaped_Obj > 40]

# Note: Direct assignment to filtered_view like filtered_view[0] = 0
# would not work as expected because filtered_view is a flattened array.
# To modify elements in the *original* array through a boolean mask,
# you apply the mask directly to the original array.

# Let's say we want to find a specific element in reshaped_Obj using filtering
# and then modify it.

# Find the coordinates of values greater than 40
rows, cols = np.where(reshaped_Obj > 40)

# Choose an element from these coordinates to modify, for example, the first one found.
if len(rows) > 0:
    target_row, target_col = rows[0], cols[0]

    print(f"\nModifying element at reshaped_Obj[{target_row}, {target_col}] which is currently {reshaped_Obj[target_row, target_col]}")

    # Modify this element through the original array reference
    reshaped_Obj[target_row, target_col] = 1000
    print("\nreshaped_Obj after modification:\n", reshaped_Obj)

    # Now let's try to achieve the same effect using a direct view filter
    # This is more direct for modifying specific elements that meet a condition
    print("\nLet's create a new original array to demonstrate direct boolean indexing modification.")
    new_original_obj = np.random.randint(1, 51, size=(4, 6))
    print("New Original Array:\n", new_original_obj)

    # Select elements greater than 40 and set them to 99
    print("\nModifying elements > 40 in new_original_obj to 99 using boolean indexing.")
    new_original_obj[new_original_obj > 40] = 99

    print("\nNew Original Array after boolean indexing modification:\n", new_original_obj)
else:
    print("No elements greater than 40 found in reshaped_Obj to demonstrate modification.")
```
