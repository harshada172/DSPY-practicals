### **Step 1: Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
import io

try:
    print("Please upload the 'mental_health.csv' file:")
    uploaded = files.upload()

    if 'mental_health.csv' in uploaded:
        df = pd.read_csv(io.BytesIO(uploaded['mental_health.csv']))
        print("Dataset 'mental_health.csv' loaded successfully.")
        df.head()
    else:
        print("Warning: 'mental_health.csv' was not among the uploaded files. Please ensure you upload the correct file.")
        df = None # Ensure df is not defined if file not found
except Exception as e:
    print(f"An error occurred during file upload or reading: {e}")
    df = None
```

## **NON-GRAPHICAL ANALYSIS**
### **1. Univariate Analysis (Stress Level)**

```python
df['Stress_Level'].describe()
```

**Objective:** To understand stress distribution

**Observation:** Mean, min, max values

**Conclusion:** Most people have moderate stress

### **2. Multivariate Analysis**

```python
df.groupby('Gender')['Stress_Level'].mean()
```

**Objective:** Compare stress across gender

**Observation:** Which gender has higher stress

**Conclusion:** One group shows higher stress

## **GRAPHICAL ANALYSIS**
### **1. Histogram (Sleep Hours)**

```python
plt.hist(df['Sleep_Hours'])
plt.title('Sleep Hours Distribution')
plt.xlabel('Sleep Hours')
plt.ylabel('Frequency')
plt.show()
```

Most people sleep between X–Y hours

Data may be normal/skewed

### **2. Scatter Plot (Screen Time vs Stress)**

```python
plt.scatter(df['Daily_Screen_Time'], df['Stress_Level'])
plt.xlabel('Screen Time')
plt.ylabel('Stress Level')
plt.show()
```

Positive relationship between screen time & stress

Higher screen time → higher stress
